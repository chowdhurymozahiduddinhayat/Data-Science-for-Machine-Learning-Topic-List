```python
import pandas as pd
from sklearn import tree
```

```python
df = pd.read_csv('game data.csv')
```

```python
df.head()
```

```python
x = df.drop('Wear Jacket?',axis=1)
```

```python
y = df['Wear Jacket?']
```

```python
x.head()
```

```python
y.head()
```

# Pandas Core Dtype Teke Is numernic dtype pra-1

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for hello in x.columns:
    if is_numeric_dtype(x[hello]):
        continue
    x[hello] = le.fit_transform(x[hello])    
```

```python
x.head()
```
